```python
print("shreyasgowdac-319_LAB-CIE")
class ModelBasedVacuumAgent:
    def __init__(self):
        self.model = {
            'Left': {'status': 'Dirty', 'clean_count': 0},
            'Right': {'status': 'Dirty', 'clean_count': 0}
        }
        self.location = 'Left'

    def perceive(self):
        """Get percepts from user input."""
        location = input("Enter current location (Left/Right): ").capitalize()
        status = input(f"Enter status of {location} (Clean/Dirty): ").capitalize()

        self.model[location]['status'] = status
        self.location = location

        return location, status

    def decide_action(self, location, status):
        """Decide action based on percept and model memory."""
        if status == 'Dirty':
            if self.model[location]['clean_count'] < 2:
                action = f"Clean {location}"
                self.model[location]['clean_count'] += 1
                self.model[location]['status'] = 'Clean'
            else:
                action = f"Skip cleaning {location} (cleaned twice already)"
        else:
            self.location = 'Right' if location == 'Left' else 'Left'
            action = f"Move to {self.location}"

        return action

    def run(self, steps=10):
        """Run the agent for a number of steps."""
        for step in range(steps):
            print(f"\n--- Step {step + 1} ---")
            location, status = self.perceive()
            action = self.decide_action(location, status)
            print(f"Action: {action}")
            print(f"Model memory: {self.model}")
            print('-' * 60)


if __name__ == "__main__":
    agent = ModelBasedVacuumAgent()
    steps = int(input("Enter number of steps to simulate: "))
    agent.run(steps)
```
